# Lesson topic: Time series. Скользящие оконные функции

### Домашнее задание

### Задача 1: Проделайте аналогичные действия, описанные в 18 lesson, для акций Корпорации Майкрософт ("MSFT")

### Задача 2: Проделайте аналогичные действия, описанные в 18 lesson, для акций Exxon Mobil Corporation ("XOM")

```python
close_px_all = pd.read_csv("stock_px.csv",
                           parse_dates=True, index_col=0)
close_px = close_px_all[["MSFT", "XOM"]]
close_px = close_px.resample("B").ffill()
#%matplotlib inline 
```

```python
close_px.plot()
```

```python
close_px.rolling(250).mean().plot() 
```

```python
std250 = close_px.pct_change().rolling(250, min_periods=10).std()
std250.plot()
```

```python
close_px.rolling(60).mean().plot(logy=True) 
```

```python
aapl_px = close_px["MSFT"]["2005":"2007"]
ma30 = aapl_px.rolling(30, min_periods=20).mean()
ewma30 = aapl_px.ewm(span=30).mean()

aapl_px.plot(style="b-", label="Price")           # style="b-"  - голубая сплошная линия
ma30.plot(style="g--", label="Simple Moving Avg") # style="g--" - зелёная прерывистая линия
ewma30.plot(style="r-", label="EW MA")            # "r-"        - красная сплошная линия
plt.legend()
```

```python
aapl_px = close_px["XOM"]["2005":"2007"]
ma30 = aapl_px.rolling(30, min_periods=20).mean()
ewma30 = aapl_px.ewm(span=30).mean()

aapl_px.plot(style="b-", label="Price")           # style="b-"  - голубая сплошная линия
ma30.plot(style="g--", label="Simple Moving Avg") # style="g--" - зелёная прерывистая линия
ewma30.plot(style="r-", label="EW MA")            # "r-"        - красная сплошная линия
plt.legend()
```

### Домашнее задание

### Задача 1: Проделайте аналогичные действия, описанные в 19 lesson, для акций Корпорации Майкрософт ("MSFT")

### Задача 2: Проделайте аналогичные действия, описанные в 19 lesson, для акций Exxon Mobil Corporation ("XOM")

```python
msft_px = close_px_all["MSFT"]
msft_rets = msft_px.pct_change()
returns = close_px_all[["XOM", "AAPL", "SPX"]].pct_change()
corr = returns["XOM"].rolling(125, min_periods=100).corr(msft_rets)
corr.plot()
```

```python
corr = returns.rolling(125, min_periods=100).corr(msft_rets)
corr.plot()
```

```python
from scipy.stats import percentileofscore
def score_at_2percent(x):
    return percentileofscore(x, 0.02)
result = returns.rolling(250).apply(score_at_2percent)
result.plot()
```
